```python
 #Is there a significant difference in median value of houses bounded by the Charles river or not?
```

```python
#We perform a T-test
```

```python
#Null Hypothesis H0: We assume both MEDVs are equal.
#Alternate Hypothesis H1: Both MEDVs are not equal.
```

```python
boston_df['CHAS'].value_counts()
```

```python
x = boston_df[boston_df['CHAS']==0]['MEDV']
x
```

```python
y = boston_df[boston_df['CHAS']==1]['MEDV']
y
```

```python
scipy.stats.ttest_ind(x,y,axis=0,equal_var=True)
```

```python
#p-value is less than alpha so null hypothesis is true.
```

```python
#Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)?
```

```python
#We perform ANOVA test for this
```

```python
boston_df['AGE'].value_counts()
```

```python
boston_df
```

```python
#Let meu1,meu2,meu3 be three different means for theree age buckets.
#null hypothesis is that all of them are equal.
```

```python
low = boston_df[boston_df['Age_group'] == '35 Years and Younger']['MEDV']
mid = boston_df[boston_df['Age_group'] == 'Between 35 to 70 years']['MEDV']
high= boston_df[boston_df['Age_group'] == '70 Years and Older']['MEDV']


f_stats, p_value = scipy.stats.f_oneway(low,mid,high,axis=0)
print("F-Statistic={0}, P-value={1}".format(f_stats,p_value))
```

#p value is more that alpha so null hypothesis selected.

```python
#Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? 
pearson,p_value = scipy.stats.pearsonr(boston_df["NOX"],boston_df["INDUS"])
print("Pearson Coefficient value={0}, P-value={1}".format(pearson,p_value))
```

```python
#we reject null hypothesis and state that there exists a relation between the two.
```
